# Tool create (Currency Conversion)

```python

from langchain_core.tools import InjectedToolArg
from typing import Annotated

@tool
def get_conversion_factor(base_currency: str, target_currency: str) -> float:
  """
  This function fetches the currency conversion factor between a given base currency and a target currency
  """
  url = f'https://v6.exchangerate-api.com/v6/c754eab14ffab33112e380ca/pair/{base_currency}/{target_currency}'

  response = requests.get(url)

  return response.json()

@tool
def convert(base_currency_value: int, conversion_rate: Annotated[float, InjectedToolArg]) -> float:
  """
  given a currency conversion rate this function calculates the target currency value from a given base currency value
  """
 
  return base_currency_value * conversion_rate
# def convert(base_currency_value: int, conversion_rate: float) -> float:
#   """
#   given a currency conversion rate this function calculates the target currency value from a given base currency value
#   """

#   return base_currency_value * conversion_rate
```

```python
convert.args
```

```python
get_conversion_factor.invoke({'base_currency':'USD','target_currency':'INR'})
```

```python
convert.invoke({'base_currency_value':10, 'conversion_rate':90.9589})
```

```python
# tool binding
llm = ChatGoogleGenerativeAI(model= 'gemini-2.5-flash-lite')
```

```python
llm_with_tools = llm.bind_tools([get_conversion_factor, convert])
```

```python
messages = [HumanMessage('What is the conversion factor between INR and USD, and based on that can you convert the 10 usd into inr')]
```

```python
messages
```

```python
ai_message = llm_with_tools.invoke(messages)
```

```python
messages.append(ai_message)
```

```python
ai_message
```

```python
ai_message.tool_calls
```

```python
import json

for tool_call in ai_message.tool_calls:
  # execute the 1st tool and get the value of conversion rate
  if tool_call['name'] == 'get_conversion_factor':
    tool_message1 = get_conversion_factor.invoke(tool_call)
    # fetch this conversion rate
    conversion_rate = json.loads(tool_message1.content)['conversion_rate']
    # append this tool message to messages list
    messages.append(tool_message1)
  # execute the 2nd tool using the conversion rate from tool 1
  if tool_call['name'] == 'convert':
    # fetch the current arg & inject the another argument which is conversion_rate
    tool_call['args']['conversion_rate'] = conversion_rate
    tool_message2 = convert.invoke(tool_call)
    messages.append(tool_message2)

```

```python
messages
```

```python
llm_with_tools.invoke(messages).content
```

```python
from langchain.agents import create_agent
# Step 5: Initialize the Agent ---
llm = ChatGoogleGenerativeAI(model= 'gemini-2.5-flash-lite')
agent_executor = create_agent(
    tools=[get_conversion_factor, convert],
    model=llm
)
```

```python
# --- Step 6: Run the Agent ---
messages = [HumanMessage(content="hi how are you?")]

response = agent_executor.invoke({"messages": messages})
print(response)
```
